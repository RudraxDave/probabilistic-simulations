# Q-1: Coin Flip Simulations
Q1-A: Basic Coin Flip Simulation

```python
import random  # Import the random module to generate random numbers

exp = 50  # Number of coin flip trials
p = 0.7  # Probability of getting heads (biased coin)

max_heads = 0  # Variable to store the longest streak of heads
curr_heads = 0  # Counter for the current streak of heads
num_heads = 0  # Counter for the total number of heads

# Loop through each trial
for i in range(exp):
    # Check if the random value is less than the probability of heads
    if random.random() < p:
        num_heads += 1  # Increment total heads count
        curr_heads += 1  # Increment current streak count
    else:
        # Update max_heads if the current streak is the longest so far
        max_heads = max(max_heads, curr_heads)
        curr_heads = 0  # Reset the current streak count

# Output the results of the simulation
print('Number of heads in 50 trials is:', num_heads)
print('Longest run of heads in 50 trials is:', max_heads)
```


Q1-B: Multiple Simulations with Varying Iterations


```python
import random  # Import the random module for generating random numbers
import matplotlib.pyplot as plt  # Import matplotlib for plotting

exp = 50  # Number of coin flip trials in each experiment
p = 0.7  # Probability of getting heads (biased coin)
iters = [20, 100, 200, 1000]  # List of different iteration counts for the experiments

# Loop through each iteration count
for k in iters:
    num_heads = [0] * k  # Initialize a list to store the number of heads for each experiment

    # Run k experiments
    for j in range(k):
        # Perform 50 trials in each experiment
        for i in range(exp):
            # Check if the random value is less than the probability of heads
            if random.random() < p:
                num_heads[j] += 1  # Increment the heads count for this experiment

    # Plot the histogram of the number of heads for this iteration count
    plt.hist(num_heads)
    plt.xlabel("Number of Heads")  # X-axis label
    plt.ylabel("Occurrences")  # Y-axis label
    plt.title(f'Q1-B: Distribution of Heads (Iterations: {k})')  # Title indicating iteration count
    plt.show()  # Display the histogram
```

Q1-C: Distribution of Head Streak Lengths

```python
import random  # Import the random module for generating random numbers
import matplotlib.pyplot as plt  # Import matplotlib for plotting

exp = 500  # Number of coin flip trials
p = 0.7  # Probability of getting heads (biased coin)

curr_heads = 0  # Counter for the current streak of heads
hist_data = []  # List to store the lengths of head streaks

# Loop through each trial
for i in range(exp):
    # Check if the random value is less than the probability of heads
    if random.random() < p:
        curr_heads += 1  # Increment current streak count
    else:
        hist_data.append(curr_heads)  # Add the current streak length to the list
        curr_heads = 0  # Reset the current streak count

# Ensure the last streak is recorded if the loop ends with a streak in progress
if curr_heads != 0:
    hist_data.append(curr_heads)

# Plot the histogram of streak lengths
plt.hist(hist_data)
plt.xlabel("Heads Run Lengths")  # X-axis label
plt.ylabel("Occurrences")  # Y-axis label
plt.title('Q1-C: Distribution of Head Streak Lengths')  # Title of the plot
plt.show()  # Display the histogram
```

Q-2: Summation of Uniform Random Variables

```python
import random  # Import the random module for generating random numbers
import matplotlib.pyplot as plt  # Import matplotlib for plotting

R = [100, 1000, 10000]  # Different numbers of realizations for the experiment
N = []  # List to store the number of draws needed to exceed the sum threshold

# Loop through each realization count
for r in R:
    for j in range(r):  # Repeat the process for each realization
        n = 1  # Start with one draw
        while True:  # Infinite loop until the condition is met
            sum = 0  # Initialize sum for this realization

            # Sum n random values
            for i in range(n):
                sum += random.random()  # Add a random number between 0 and 1 to the sum

            # Check if the sum exceeds 4
            if sum > 4:
                break  # Exit the loop if the sum exceeds the threshold
            n += 1  # Increment the draw count

        N.append(n)  # Store the number of draws needed for this realization

    # Plot the histogram of the number of draws needed
    plt.hist(N, density=True, bins=10)
    plt.xlabel("Number of Standard Uniform RVs")  # X-axis label
    plt.ylabel("Occurrences")  # Y-axis label
    plt.title(f'Q2: Summation of Uniform RVs (Realizations: {r})')  # Title indicating realization count
    plt.show()  # Display the histogram
```

# Additional Experiment 1: Simulating the Law of Large Numbers
Description: This experiment will simulate the Law of Large Numbers, which states that as the number of trials increases, the sample mean will converge to the expected value.

```python
import random  # Import the random module for generating random numbers
import matplotlib.pyplot as plt  # Import matplotlib for plotting

p = 0.7  # Probability of getting heads (biased coin)
trials = [10, 100, 1000, 10000]  # Different numbers of trials

# Initialize a list to store the sample means for each number of trials
means = []

# Loop through each trial count
for t in trials:
    num_heads = 0  # Counter for the number of heads
    # Perform t trials
    for i in range(t):
        if random.random() < p:  # Check if the random value is less than the probability of heads
            num_heads += 1  # Increment the heads count
    sample_mean = num_heads / t  # Calculate the sample mean
    means.append(sample_mean)  # Store the sample mean

# Plot the sample means
plt.plot(trials, means, marker='o')
plt.axhline(y=p, color='r', linestyle='--')  # Add a horizontal line for the expected value
plt.xscale('log')  # Use a logarithmic scale for the x-axis
plt.xlabel("Number of Trials")  # X-axis label
plt.ylabel("Sample Mean")  # Y-axis label
plt.title('Law of Large Numbers')  # Title of the plot
plt.show()  # Display the plot
```

# Additional Experiment 2: Estimating Pi Using a Monte Carlo Method
Description: This experiment will estimate the value of  (Pi) using a Monte Carlo method, where random points are plotted inside a square and the ratio of points inside a quarter circle is used to approximate Pi.

```python
import random  # Import the random module for generating random numbers
import matplotlib.pyplot as plt  # Import matplotlib for plotting

total_points = 10000  # Number of random points
inside_circle = 0  # Counter for points inside the quarter circle

x_inside = []  # List to store x-coordinates of points inside the quarter circle
y_inside = []  # List to store y-coordinates of points inside the quarter circle
x_outside = []  # List to store x-coordinates of points outside the quarter circle
y_outside = []  # List to store y-coordinates of points outside the quarter circle

# Generate random points
for i in range(total_points):
    x = random.random()  # Random x-coordinate between 0 and 1
    y = random.random()  # Random y-coordinate between 0 and 1
    if x**2 + y**2 <= 1:  # Check if the point is inside the quarter circle
        inside_circle += 1  # Increment the inside circle count
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)

# Estimate Pi
pi_estimate = (inside_circle / total_points) * 4
print(f"Estimated value of Pi: {pi_estimate}")

# Plot the points
plt.scatter(x_inside, y_inside, color='blue', s=1, label='Inside Circle')
plt.scatter(x_outside, y_outside, color='red', s=1, label='Outside Circle')
plt.xlabel("X")  # X-axis label
plt.ylabel("Y")  # Y-axis label
plt.title('Estimating Pi using Monte Carlo Method')  # Title of the plot
plt.legend()
plt.show()  # Display the plot
```

# Additional Experiment 3: Random Walk Simulation
Description: This experiment simulates a 1D random walk, where at each step, the position either increases or decreases by one unit with equal probability.

```python
import random  # Import the random module for generating random numbers
import matplotlib.pyplot as plt  # Import matplotlib for plotting

steps = 1000  # Number of steps in the random walk
position = 0  # Starting position
positions = [position]  # List to store the positions at each step

# Simulate the random walk
for _ in range(steps):
    step = random.choice([-1, 1])  # Randomly choose to step left (-1) or right (+1)
    position += step  # Update the position
    positions.append(position)  # Store the new position

# Plot the random walk
plt.plot(positions)
plt.xlabel("Steps")  # X-axis label
plt.ylabel("Position")  # Y-axis label
plt.title('1D Random Walk Simulation')  # Title of the plot
plt.show()  # Display the plot
```

# Additional Experiment 4: Buffon's Needle Problem
Description: This classic problem estimates the value of  by simulating the dropping of a needle on a floor with parallel lines and counting the number of times the needle crosses a line.

```python
import random  # Import the random module for generating random numbers
import math  # Import math for trigonometric functions

needle_length = 1  # Length of the needle
line_spacing = 2  # Distance between lines on the floor
trials = 10000  # Number of needle drops
crosses = 0  # Counter for the number of times the needle crosses a line

# Simulate the needle drops
for _ in range(trials):
    center_distance = random.uniform(0, line_spacing / 2)  # Distance from needle center to the nearest line
    angle = random.uniform(0, math.pi / 2)  # Angle of the needle relative to the lines
    if center_distance <= (needle_length / 2) * math.sin(angle):  # Check if the needle crosses a line
        crosses += 1  # Increment the crosses count

# Estimate Pi using the ratio of crosses to total drops
pi_estimate = (2 * needle_length * trials) / (crosses * line_spacing)
print(f"Estimated value of Pi using Buffon's Needle: {pi_estimate}")
```
